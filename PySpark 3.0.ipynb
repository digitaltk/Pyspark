```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('Calc').getOrCreate()
```

```python
#Dropping Null values
df_pyspark=spark.read.csv('MOCK_DATAE.csv',header=True,inferSchema=True)
```

```python
df_pyspark.show()
```

```python
df_pyspark.drop('last_name')
```

```python
df_pyspark.na.drop()
```

```python
###any==how All thevlaues are null only you can get 
df_pyspark.na.drop(how='all')
```

```python
##threshold if more than 2 null values
df_pyspark.na.drop(how='all',thresh=2)
```

```python
##Subset drop from specfic columns
df_pyspark.na.drop(how='all',subset=['ip_address'])
```

```python
##Missing values 
df_pyspark.na.fill('Missing Values')
```

```python
df_pyspark.na.fill('Missing Values',['last_name','email']).show()
```

```python
##mean mode median
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['id'],
    outputCols=["{}_imputed".format(c) for c in ['id']]
).setStrategy("mean")
```

```python
imputer.fit(df_pyspark).transform(df_pyspark)
```
